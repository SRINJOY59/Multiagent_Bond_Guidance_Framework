```python
import pandas as pd
import json
import re


df['Coupon_Rate'] = None
df['Face_Value'] = None
df['Total_Issue_Size'] = None
df['Redemption_Date'] = None
df['Instrument_Name'] = None
df['Listing_Details'] = None

def extract_coupon_rate(desc):
    if desc and isinstance(desc, str):
        match = re.search(r'(\d+\.\d+)%', desc)
        if match:
            return f"{match.group(1)}%"
    return None

for i in range(len(df)):
    try:
        # Parse the JSON string to a dictionary
        if pd.notna(df['instrument_details'][i]):
            instrument_data = json.loads(df['instrument_details'][i])
            
            # Extract instruments data if it exists
            if 'instrumentsVo' in instrument_data and 'instruments' in instrument_data['instrumentsVo']:
                instruments = instrument_data['instrumentsVo']['instruments']
                
                # Extract coupon rate from instrument description
                coupon_rate = None
                if 'instrumentDesc' in instruments:
                    coupon_rate = extract_coupon_rate(instruments['instrumentDesc'])
                
                # Extract face value
                face_value = None
                if 'faceValue' in instruments and instruments['faceValue'] is not None:
                    face_value = f"₹{instruments['faceValue']:,}"
                
                # Extract total issue size
                issue_size = None
                if 'totalIssueSize' in instruments and instruments['totalIssueSize'] not in [None, '-']:
                    try:
                        issue_size = f"₹{float(instruments['totalIssueSize']):,.2f} Cr"
                    except (ValueError, TypeError):
                        issue_size = instruments['totalIssueSize']
                
                # Extract redemption date
                redemption_date = instruments.get('redemptionDate')
                
                # Create instrument name
                instrument_name = None
                if coupon_rate and 'secured' in instruments:
                    secured_status = "Secured" if instruments['secured'] == "Secured" else "Unsecured"
                    instrument_name = f"{coupon_rate} RIL {secured_status} NCDs"
                
                # Extract listing details
                listing_details = None
                if 'instrumentDesc' in instruments:
                    if 'LISTED' in instruments['instrumentDesc'].upper():
                        listing_details = "Listed"
                    else:
                        listing_details = "Not Listed"
                
                # Update the DataFrame with extracted values
                df.at[i, 'Coupon_Rate'] = coupon_rate
                df.at[i, 'Face_Value'] = face_value
                df.at[i, 'Total_Issue_Size'] = issue_size
                df.at[i, 'Redemption_Date'] = redemption_date
                df.at[i, 'Instrument_Name'] = instrument_name
                df.at[i, 'Listing_Details'] = listing_details
                
    except (json.JSONDecodeError, TypeError) as e:
        print(f"Error processing row {i}: Invalid JSON format or None value - {str(e)}")
    except Exception as e:
        print(f"Error processing row {i}: {str(e)}")
```

```python
import pandas as pd
import json
from datetime import datetime

# Assuming df is your DataFrame with a 'rating_details' column containing JSON strings

def extract_credit_rating(json_str):
    try:
        if pd.isna(json_str):
            return None
            
        rating_data = json.loads(json_str)
        
        # Check if there are current ratings
        if 'currentRatings' in rating_data and rating_data['currentRatings']:
            # Get the most recent rating based on creditRatingDate
            current_ratings = rating_data['currentRatings']
            
            # Convert dates to datetime objects for comparison (ignoring invalid dates)
            dated_ratings = []
            for rating in current_ratings:
                try:
                    if rating['creditRatingDate'] and rating['creditRatingDate'] != '-':
                        date = datetime.strptime(rating['creditRatingDate'], '%d-%m-%Y')
                        dated_ratings.append((date, rating))
                except (ValueError, TypeError):
                    continue
            
            # Sort by date (most recent first)
            dated_ratings.sort(reverse=True)
            
            if dated_ratings:
                # Use the most recent rating
                latest_rating = dated_ratings[0][1]
                rating_value = latest_rating['currentRating']
                outlook = latest_rating['outlook'] if latest_rating['outlook'] and latest_rating['outlook'] != '-' else ''
                agency = latest_rating['creditRatingAgencyName']
                
                # Format the credit rating string
                if outlook:
                    return f"{rating_value} ({outlook}) by {agency}"
                else:
                    return f"{rating_value} by {agency}"
            
        return None
        
    except (json.JSONDecodeError, TypeError):
        return None
    except Exception as e:
        print(f"Unexpected error: {str(e)}")
        return None

df['Credit_Rating'] = df['credit_rating_details'].apply(extract_credit_rating)
```

```python
def extract_key_documents(json_str):
    try:
        if pd.isna(json_str):
            return None
            
        document_data = json.loads(json_str)
        
        # Dictionary to store important document types and their links
        key_docs = {}
        
        # Regex patterns to identify document types
        offer_doc_pattern = re.compile(r'prospectus|memorandum|offer|im\b', re.IGNORECASE)
        trust_deed_pattern = re.compile(r'trust|deed', re.IGNORECASE)
        
        # Process each document entry
        for doc in document_data:
            doc_name = doc.get('docName', '')
            download_link = doc.get('downloadLink', '')
            
            if offer_doc_pattern.search(doc_name):
                key_docs['Offer Document'] = download_link
            elif trust_deed_pattern.search(doc_name):
                key_docs['Trust Deed'] = download_link
        
        # If specific document types weren't found, include other important documents
        if len(key_docs) < 2:
            # Sort documents by possible importance (prioritizing BSE/NSE links)
            bse_nse_docs = [doc for doc in document_data if 'bse' in doc.get('downloadLink', '').lower() 
                           or 'nse' in doc.get('downloadLink', '').lower()]
            
            if bse_nse_docs and 'Offer Document' not in key_docs:
                key_docs['Offer Document'] = bse_nse_docs[0].get('downloadLink', '')
            
            # Add rating documents if needed
            rating_docs = [doc for doc in document_data if 'rating' in doc.get('docName', '').lower()]
            if rating_docs and len(key_docs) < 2:
                key_docs['Rating Document'] = rating_docs[0].get('downloadLink', '')
        
        # Format the key documents as requested
        formatted_docs = []
        for doc_type, link in key_docs.items():
            formatted_docs.append(f"[{doc_type} PDF]({link})")
        
        # Join the formatted documents with commas
        return ', '.join(formatted_docs) if formatted_docs else None
        
    except (json.JSONDecodeError, TypeError):
        return None
    except Exception as e:
        print(f"Unexpected error: {str(e)}")
        return None

# Apply the function to extract key documents
df['Key_Documents'] = df['key_documents_details'].apply(extract_key_documents)
```
